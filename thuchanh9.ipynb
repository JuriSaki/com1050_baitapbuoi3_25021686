```python
def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1
    return merged_list

list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
merged = merge_sorted_lists(list1, list2)
print(merged)
```

```python
def count_less_equal(matrix, mid, n, m):
    count = 0
    row = n - 1
    col = 0
    while row >= 0 and col < m:
        if matrix[row][col] <= mid:
            count += (row + 1)
            col += 1
        else:
            row -= 1
    return count

def find_kth_smallest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    low = matrix[0][0]
    high = matrix[n-1][m-1]

    ans = low
    while low <= high:
        mid = (low + high) // 2
        if count_less_equal(matrix, mid, n, m) >= k:
            ans = mid
            high = mid - 1
        else:
            low = mid + 1
    return ans

def main():
    try:
        n, m, k = map(int, input().split())

        matrix = []
        for _ in range(n):
            matrix.append(list(map(int, input().split())))
        target_k = (n * m) - k + 1

        print(find_kth_smallest(matrix, target_k))

    except (EOFError, ValueError):
        pass

if __name__ == "__main__":
    main()
```

```python
def solve_smallest_swap(n_str):
    s = list(str(n_str))
    length = len(s)
    min_num = None
    for i in range(length):
        for j in range(i + 1, length):
            temp = s.copy()
            temp[i], temp[j] = temp[j], temp[i]
            current_val = int("".join(temp))
            if min_num is None or current_val < min_num:
                min_num = current_val
    return min_num
inputs = str(input()).split()

for inp in inputs:
    print(f"{solve_smallest_swap(inp)}")
```

```python
def check_is_subarray():
    try:

        line1 = input("Nhập list1: ").split()
        list1 = [int(x) for x in line1]


        line2 = input("Nhập list2: ").split()
        list2 = [int(x) for x in line2]

        n, m = len(list2), len(list1)
        found = False


        for i in range(n - m + 1):
            if list2[i : i + m] == list1:
                found = True
                break

        if found:
            print("YES")
        else:
            print("NO")

    except EOFError:
        pass


check_is_subarray()
```

```python
def canPlaceFlowers(flowerbed, n):
    count = 0
    length = len(flowerbed)

    for i in range(length):
        if flowerbed[i] == 0:
            empty_left = (i == 0) or (flowerbed[i - 1] == 0)
            empty_right = (i == length - 1) or (flowerbed[i + 1] == 0)

            if empty_left and empty_right:
                flowerbed[i] = 1
                count += 1

                if count >= n:
                    return True
    return count >= n
flowerbed = list(map(int, input().split()))
n = int(input())
print(canPlaceFlowers(flowerbed, n))
```

```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}

    for move in moves:
        if move == 'R': x += 1
        elif move == 'L': x -= 1
        elif move == 'U': y += 1
        elif move == 'D': y -= 1

        current_pos = (x, y)
        if current_pos in visited:
            return True
        visited.add(current_pos)

    return False

if __name__ == "__main__":
    while True:
        try:
            line = input()
            if not line: break
            moves = line.strip()
            print(is_path_crossing(moves))

        except EOFError:
            break
```

```python
def Median(nums):
    a, b, c, d, e = nums

    if a > b: a, b = b, a
    if c > d: c, d = d, c
    if a > c: a, b, c, d = c, d, a, b

    if b > e: b, e = e, b
    if b > c: b, e, c, d = c, d, b, e

    if c < e: return c
    return e

if __name__ == "__main__":
    while True:
        try:
            line = input()
            if not line: break
            nums = list(map(int, line.split()))
            if len(nums) == 5:
                print(Median(nums))
        except EOFError:
            break
```

```python
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    radius = 0
    i = 0
    n = len(heaters)

    for h in houses:
        while i < n - 1 and abs(heaters[i+1] - h) <= abs(heaters[i] - h):
            i += 1
        radius = max(radius, abs(heaters[i] - h))

    return radius

if __name__ == "__main__":
    while True:
        try:
            line_houses = input()
            if not line_houses: break
            houses = list(map(int, line_houses.split()))

            line_heaters = input()
            heaters = list(map(int, line_heaters.split()))

            print(findRadius(houses, heaters))
        except EOFError:
            break
        except ValueError:
            break
```
