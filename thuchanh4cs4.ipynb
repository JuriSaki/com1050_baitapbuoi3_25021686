```python
#W4A8
n = int(input(" Enter a number "))
def reverse(n):
    kq=0
    count=10**(len(str(n))-1)
    while n>0:
        kq=kq+count*(n%10)
        n//=10
        count/=10
    return int(kq)
def palin(n):
    count = 0
    while n > 0:
        n = n + reverse(n)
        count += 1
        if n == reverse(n):
            return (count, n)
print(palin(n))
```

```python
#W4A9
def distinct_digits(num):
    digits = set()
    while num > 0:
        d = num % 10
        if d in digits:
            return False
        digits.add(d)
        num //= 10
    return True
n = int(input("Enter a number: "))
results = []
i = 1
while i * i <= n:
    sq = i * i
    if distinct_digits(sq):
        results.append(sq)
    i += 1

if results:
    print(" ".join(map(str, results)))
else:
    print("(no number)")
```

```python
#W4A10
def collatz_length(n):
    length = 1
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
        length += 1
    return length

n = int(input("Enter a number: "))

max_len = 0
best_x = 1

for x in range(1, n+1):
    L = collatz_length(x)
    if L > max_len or (L == max_len and x < best_x):
        max_len = L
        best_x = x

print(best_x, max_len)
```

```python
#W4A13
def tong_uoc(n):
    tong = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            tong += i
            if i != n // i:
                tong += n // i
    return tong
a = int(input("enter the first number: "))
b = int(input("enter the second number: "))
if tong_uoc(a) == b and tong_uoc(b) == a:
    print("true")
else:
    print("false")
```

```python
#W4A27
s = input("Nhập chuỗi: ")
def my_len(st):
    count = 0
    for _ in st:
        count += 1
    return count

n = my_len(s)

count_words = 0
in_word = False

for i in range(n):
    if s[i] != " " and not in_word:
        count_words += 1
        in_word = True
    elif s[i] == " ":
        in_word = False

print("Số từ trong chuỗi là:", count_words)
```

```python
#W4A30
def dem_ky_tu(chuoi):
    hoa = 0
    thuong = 0
    so = 0
    for c in chuoi:
        if c.isupper():
            hoa += 1
        elif c.islower():
            thuong += 1
        elif c.isdigit():
            so += 1
    return hoa, thuong, so
chuoi = input("Nhập chuỗi: ")
hoa, thuong, so = dem_ky_tu(chuoi)
print("Số ký tự in hoa:", hoa)
print("Số ký tự in thường:", thuong)
print("Số ký tự số:", so)
```

```python
#W4A32
import string

def kiem_tra_mat_khau(password):
    co_hoa = any(c.isupper() for c in password)
    co_thuong = any(c.islower() for c in password)
    co_so = any(c.isdigit() for c in password)
    co_dac_biet = any(c in string.punctuation for c in password)
    du_dai = len(password) > 6
    if co_hoa and co_thuong and co_so and co_dac_biet and du_dai:
        return True
    else:
        return False
chuoi = input("Nhập mật khẩu: ")
if kiem_tra_mat_khau(chuoi):
    print("Đây là mật khẩu mạnh")
else:
    print("Mật khẩu chưa đủ mạnh.")
```

```python
#W4A34
def xoa_chuoi(a, b):
    ket_qua = ""
    i = 0
    while i < len(a):
        if a[i:i+len(b)] == b:
            i += len(b)
        else:
            ket_qua += a[i]
            i += 1
    return ket_qua
a = input("Nhập chuỗi a: ")
b = input("Nhập chuỗi b: ")
ket_qua = xoa_chuoi(a, b)
print("Chuỗi sau khi xóa:", ket_qua)
```
