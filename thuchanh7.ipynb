```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
arr = list(map(int, input().split()))
target = int(input())
result = binary_search(arr, target)
print(result)
```

```python
def count_occurrences(arr, target):
    count = 0
    for num in arr:
        if num == target:
            count += 1
    return count
arr = list(map(int, input().split()))
target = int(input())
result = count_occurrences(arr, target)
print(result)
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
arr = list(map(int, input().split()))
sorted_arr = bubble_sort(arr)
print(sorted_arr)
```

```python
def most_appearances(arr):
    frequency = {}
    for num in arr:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1
    max_count = max(frequency.values())
    most_frequent = [num for num, count in frequency.items() if count == max_count]
    if len(most_frequent) > 1:
        most_frequent = most_frequent[0]
    return most_frequent, max_count
arr = list(map(int, input().split()))
most_frequent, count = most_appearances(arr)
print(most_frequent, count)
```

```python
def count_pairs(arr, X):
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == X:
                count += 1
    return count
arr = list(map(int, input().split()))
X = int(input())
print(count_pairs(arr, X))
```

```python
def flatten_list(nested_list):
    flat = []
    for sublist in nested_list:
        for item in sublist:
            flat.append(item)
    return flat
nested = list(eval(input()))
print(flatten_list(nested))
```

```python
def length_of_LIS(arr):
    n = len(arr)
    if n == 0:
        return 0
    lis = [1] * n
    for i in range(1, n):
        for j in range(0, i):
            if arr[i] > arr[j] and lis[i] < lis[j] + 1:
                lis[i] = lis[j] + 1
    return max(lis)
arr = eval(input())
print(length_of_LIS(arr))
```

```python
import ast
import bisect

A = ast.literal_eval(input())
B = ast.literal_eval(input())
X = []

for i in range (len(A)):
    X.append((A[i][1] - A[i][0] + 1,A[i][0],A[i][1]))

X.sort()
B.sort()

Ans = [0] * len(B)

for i in range (len(X) - 1, -1, -1):
    x1 = bisect.bisect_left (B, X[i][1])
    x2 = bisect.bisect_right (B, X[i][2])
    for j in range (x1, x2): Ans[j] = X[i][0]

for i in range (len(B)):
    if Ans[i] == 0: Ans[i] = -1

print(Ans)
```

```python
def count_smaller(nums):
    n = len(nums)
    result = [0] * n
    enum_nums = list(enumerate(nums))
    items = [(v, i) for i, v in enum_nums]

    def merge_sort(arr):
        if len(arr) <= 1:
            return arr

        mid = len(arr) // 2
        left = merge_sort(arr[:mid])
        right = merge_sort(arr[mid:])

        return merge(left, right)

    def merge(left, right):
        merged = []
        i = j = 0

        while i < len(left) and j < len(right):
            if left[i][0] > right[j][0]:
                merged.append(right[j])
                j += 1
            else:
                result[left[i][1]] += j
                merged.append(left[i])
                i += 1
        while i < len(left):
            result[left[i][1]] += j
            merged.append(left[i])
            i += 1
        while j < len(right):
            merged.append(right[j])
            j += 1

        return merged

    merge_sort(items)
    return result
nums = eval(input())
print(count_smaller(nums))
```

```python
import bisect

def max_subarray_mod(nums, m):
    max_mod = 0
    current_prefix_sum = 0
    prefix_sums = []

    for x in nums:
        current_prefix_sum = (current_prefix_sum + x) % m
        max_mod = max(max_mod, current_prefix_sum)
        idx = bisect.bisect_right(prefix_sums, current_prefix_sum)

        if idx < len(prefix_sums):
            max_mod = max(max_mod, (current_prefix_sum - prefix_sums[idx] + m) % m)
        bisect.insort(prefix_sums, current_prefix_sum)

    return max_mod

nums = eval(input())
m = int(input())
print(max_subarray_mod(nums, m))
```

```python
def day_hai_hoa(arr):
    counts = {}
    for num in arr:
        counts[num] = counts.get(num, 0) + 1
    max_length = 0
    for num in counts:
        if num + 1 in counts:
            current_length = counts[num] + counts[num + 1]
            max_length = max(max_length, current_length)

    return max_length
arr = list(map(int, input().split()))
print(day_hai_hoa(arr))
```

```python
def find_nemo_neighbors():
    try:
        n = int(input())

        guests = []
        for _ in range(n):
            guests.append(input().strip())
        try:
            nemo_index = guests.index("Nemo")
        except ValueError:
            return
        before_index = (nemo_index - 1 + n) % n
        after_index = (nemo_index + 1) % n

        name1 = guests[before_index]
        name2 = guests[after_index]
        print(f"{name1} and {name2}")

    except EOFError:
        pass
if __name__ == "__main__":
    find_nemo_neighbors()
```
