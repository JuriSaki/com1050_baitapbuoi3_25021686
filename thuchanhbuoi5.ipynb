```python
def ktnt(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input(" Enter a number: "))
while n <= 0:
    n = int(input(" Enter a positive number"))
if ktnt(n):
    print("True")
else:
    print("False")
```

```python
def is_perfect_number(n: int) -> bool:
    if n <= 1:
        return False
    tong_uoc = 0
    for i in range(1, n):
        if n % i == 0:
            tong_uoc += i
    return tong_uoc == n
n = int(input("Enter a positive integer: "))
print(is_perfect_number(n))
```

```python
def factorial(n: int) -> int:
    if n < 0:
        raise ValueError("Không tồn tại giai thừa cho số âm")
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
n = int(input("Nhập số nguyên n: "))
print(f"{n}! =", factorial(n))
```

```python
def calculator(num1: float, num2: float, operat: str) -> float:
    if operat == '+':
        result = num1 + num2
    elif operat == '-':
        result = num1 - num2
    elif operat == '*':
        result = num1 * num2
    elif operat == '/':
        if num2 == 0:
            raise ValueError("Không thể chia cho 0")
        result = num1 / num2
    else:
        raise ValueError("Toán tử không hợp lệ. Chỉ dùng +, -, *, /")

    return round(result, 2)
input_values = input("Nhập num1 num2 operat (vd: 5.5 2.3 +): ").split()
num1, num2 = map(float, input_values[:2])
operat = input_values[2]
print("Kết quả:", calculator(num1, num2, operat))
```

```python
def hamming(x, y):
    result = x ^ y
    return bin(result).count('1')
x, y = map(int, input().split())
print(hamming(x, y))
```

```python
def sum_of_digits(n: int) -> int:
    if n < 0:
        raise ValueError("Chỉ nhận số nguyên dương")
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
n = int(input("Nhập số nguyên dương n: "))
print(sum_of_digits(n))
```

```python
def dang_cau(a: str, b: str) -> bool:
    if len(a) != len(b):
        return False
    map_a_b = {}
    map_b_a = {}
    for i in range(len(a)):
        char_a = a[i]
        char_b = b[i]
        if char_a in map_a_b:
            if map_a_b[char_a] != char_b:
                return False
        else:
            map_a_b[char_a] = char_b
        if char_b in map_b_a:
            if map_b_a[char_b] != char_a:
                return False
        else:
            map_b_a[char_b] = char_a
    return True
a, b = input().split()
print(dang_cau(a, b))
```
