```python
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        self.pi = 3.14

    def surface_area(self):
        return 2 * self.pi * self.radius * (self.radius + self.height)

    def volume(self):
        return self.pi * (self.radius ** 2) * self.height
r, h = map(float, input().split())
cyl = Cylinder(r, h)
print(f"{cyl.surface_area():.2f} {cyl.volume():.2f}")
```

```python
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0)

    def is_valid(self):
        if self.month < 1 or self.month > 12:
            return False
        days_in_month = [31, 29 if self.is_leap_year() else 28, 31, 30, 31, 30,
                         31, 31, 30, 31, 30, 31]
        if self.day < 1 or self.day > days_in_month[self.month - 1]:
            return False
        return True

    def next_day(self):
        if not self.is_valid():
            return None
        days_in_month = [31, 29 if self.is_leap_year() else 28, 31, 30, 31, 30,
                         31, 31, 30, 31, 30, 31]
        d, m, y = self.day, self.month, self.year
        if d < days_in_month[m - 1]:
            d += 1
        else:
            d = 1
            if m == 12:
                m = 1
                y += 1
            else:
                m += 1
        return Date(d, m, y)
date_str = input().strip()
try:
    dd, mm, yyyy = map(int, date_str.split("/"))
    d = Date(dd, mm, yyyy)
    if not d.is_valid():
        print("INVALID")
    else:
        next_d = d.next_day()
        print(f"{next_d.day:02d}/{next_d.month:02d}/{next_d.year:04d}")
except:
    print("INVALID")
```

```python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def is_origin(self):
        return self.x == 0 and self.y == 0

    def is_on_x_axis(self):
        return self.y == 0 and self.x != 0

    def is_on_y_axis(self):
        return self.x == 0 and self.y != 0

    def distance_to(self, other):
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)
x, y = map(float, input().split())
M = Point(x, y)
O = Point(0, 0)

if M.is_origin():
    print("M là gốc tọa độ")
elif M.is_on_y_axis():
    print("M nằm trên trục tung")
elif M.is_on_x_axis():
    print("M nằm trên trục hoành")
else:
    print("M không nằm trên gốc tọa độ, trục tung, trục hoành")
dist = M.distance_to(O)
print(f"Khoảng cách từ M đến gốc tọa độ: {dist:.2f}")
```

```python
class ShoppingCart:
    def __init__(self):
        self.products = []

    def add_product(self, name, price):
        self.products.append({"name": name, "price": price})

    def remove_product(self, name):
        self.products = [p for p in self.products if p["name"] != name]

    def is_empty(self):
        if len(self.products) == 0:
            print("Giỏ hàng đang rỗng.")
            return True
        else:
            print("Giỏ hàng đang có sản phẩm.")
            return False

    def calculate_total(self):
        return sum(p["price"] for p in self.products)

    def display_all(self):
        for p in self.products:
            print(f"{p['name']}: {p['price']}")

    def clear_cart(self):
        self.products = []

def main():
    cart = ShoppingCart()

    while True:
        choice = input()

        if choice == '2':
            name = input()
            price = float(input())
            cart.add_product(name, price)

        elif choice == '3':
            name = input()
            cart.remove_product(name)

        elif choice == '4':
            cart.is_empty()

        elif choice == '5':
            print(cart.calculate_total())

        elif choice == '6':
            cart.display_all()

        elif choice == '7':
            cart.clear_cart()
        if input("Có muốn thoát khỏi chương trình không? ").lower() == "yes":
            break

if __name__ == "__main__":
    main()
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)
    def scale(self, fc):
        self.length *= fc
        self.width *= fc
h, w, fc = map(float, input().split())
rect = Rectangle(h, w)
rect.scale(fc)
print(f"{rect.area():.2f} {rect.perimeter():.2f}")
```

```python
import math

class Fraction:
    def __init__(self, num, den):
        common = math.gcd(num, den)
        self.num = num // common
        self.den = den // common
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den

    def add(self, other):
        return Fraction(self.num * other.den + other.num * self.den, self.den * other.den)

    def sub(self, other):
        return Fraction(self.num * other.den - other.num * self.den, self.den * other.den)

    def mul(self, other):
        return Fraction(self.num * other.num, self.den * other.den)

    def div(self, other):
        return Fraction(self.num * other.den, self.den * other.num)

    def __str__(self):
        return f"{self.num}/{self.den}"

def main():
    try:
        data = input().split()
        if not data: return

        a, b = int(data[0]), int(data[1])
        op = data[2]
        c, d = int(data[3]), int(data[4])

        f1 = Fraction(a, b)
        f2 = Fraction(c, d)

        if op == '+':
            print(f1.add(f2))
        elif op == '-':
            print(f1.sub(f2))
        elif op == '*':
            print(f1.mul(f2))
        elif op == '/':
            print(f1.div(f2))

    except EOFError:
        pass

if __name__ == "__main__":
    main()
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += float(x)

    def withdraw(self, x):
        amount = float(x)
        if amount <= self.balance:
            self.balance -= amount
    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"
def main():
    try:
        line1 = input().split()
        if not line1: return
        owner = line1[0]
        initial_balance = line1[1]
        account = BankAccount(owner, initial_balance)
        n = int(input())
        for _ in range(n):
            command_line = input().split()
            action = command_line[0].upper()
            x = command_line[1]

            if action == "DEPOSIT":
                account.deposit(x)
            elif action == "WITHDRAW":
                account.withdraw(x)
        print(f"{account.balance:.2f}")

    except (EOFError, ValueError, IndexError):
        pass

if __name__ == "__main__":
    main()
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(float(score))

    def avg(self):
        if not self.scores:
            return 0.0
        return sum(self.scores) / len(self.scores)

    def rank(self):
        average = self.avg()
        if average >= 8:
            return "Excellent"
        elif average >= 6.5:
            return "Good"
        elif average >= 5:
            return "Average"
        else:
            return "Poor"

def main():
    try:
        name = input().strip()
        student = Student(name)
        line_n = input().strip()
        if not line_n: return
        n = int(line_n)
        for _ in range(n):
            data = input().split()
            if len(data) >= 2:
                student.add_score(data[0], data[1])
        average_score = student.avg()
        ranking = student.rank()
        print(f"{student.name} {average_score:.2f} {ranking}")

    except (EOFError, ValueError):
        pass

if __name__ == "__main__":
    main()
```

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        count = 0
        for b in self.books:
            if b.author == name:
                count += 1
        return count

    def find_by_year(self, y):
        count = 0
        for b in self.books:
            if b.year == int(y):
                count += 1
        return count

def main():
    library = Library()
    try:
        line1 = input().strip()
        if not line1:
            return
        n = int(line1)
    except ValueError:
        return

    for _ in range(n):
        line = input().strip()
        if not line:
            continue

        if line.startswith("ADD "):
            content = line[4:]
            parts = content.split(';')
            if len(parts) == 3:
                title = parts[0]
                author = parts[1]
                year = parts[2]
                library.add(Book(title, author, year))

        elif line.startswith("COUNT "):
            author_name = line[6:]
            print(library.count_by_author(author_name))

        elif line.startswith("COUNTYEAR "):
            year_val = line[10:]
            print(library.find_by_year(year_val))

if __name__ == "__main__":
    main()
```

```python
import math

class Points:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __sub__(self, no):
        return Points(self.x - no.x, self.y - no.y, self.z - no.z)

    def dot(self, no):
        return self.x * no.x + self.y * no.y + self.z * no.z

    def cross(self, no):
        return Points(
            self.y * no.z - self.z * no.y,
            self.z * no.x - self.x * no.z,
            self.x * no.y - self.y * no.x
        )

    def absolute(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

def main():
    try:
        points = []
        for _ in range(4):
            x, y, z = map(float, input().split())
            points.append(Points(x, y, z))
        a, b, c, d = points[0], points[1], points[2], points[3]
        ab = b - a
        bc = c - b
        cd = d - c
        n1 = ab.cross(bc)
        n2 = bc.cross(cd)
        cos_phi = n1.dot(n2) / (n1.absolute() * n2.absolute())
        cos_phi = max(min(cos_phi, 1.0), -1.0)
        phi_rad = math.acos(cos_phi)
        phi_deg = math.degrees(phi_rad)
        print(f"{phi_deg:.2f}")

    except EOFError:
        pass

if __name__ == "__main__":
    main()
```

```python
class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.real = float(realPart)
        self.imag = float(imaginaryPart)

    def multiply(self, other):
        new_real = self.real * other.real - self.imag * other.imag
        new_imag = self.real * other.imag + self.imag * other.real
        return ComplexNumber(new_real, new_imag)

    def __str__(self):
        return f"{self.real:.2f} {self.imag:.2f}"

def main():
    try:
        data1 = input().split()
        if not data1: return
        c1 = ComplexNumber(data1[0], data1[1])

        data2 = input().split()
        if not data2: return
        c2 = ComplexNumber(data2[0], data2[1])

        result = c1.multiply(c2)
        print(result)

    except (EOFError, ValueError, IndexError):
        pass

if __name__ == "__main__":
    main()
```

```python
class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self.a = int(edge1)
        self.b = int(edge2)
        self.c = int(edge3)

    def getArea(self):
        if (self.a + self.b > self.c) and \
           (self.a + self.c > self.b) and \
           (self.b + self.c > self.a):

            p = (self.a + self.b + self.c) / 2
            area = (p * (p - self.a) * (p - self.b) * (p - self.c)) ** 0.5
            return area
        else:
            return -1.0

def main():
    try:
        line = input().split()

        if len(line) != 3:
            print("invalid")
            return

        a, b, c = line
        tri = Triangle(a, b, c)
        result = tri.getArea()

        if result == -1.0:
            print(result)
        else:
            print(f"{result:.2f}")

    except (EOFError, ValueError):
        print("invalid")

if __name__ == "__main__":
    main()
```
