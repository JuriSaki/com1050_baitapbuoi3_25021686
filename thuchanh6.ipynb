```python
s1 = input()
s2 = input()


def parse_dict(s):
    d = {}
    for pair in s.split():
        k, v = pair.split(":")
        d[k] = int(v)
    return d

dict1 = parse_dict(s1)
dict2 = parse_dict(s2)


merged = dict1.copy()
for k, v in dict2.items():
    if k in merged:
        merged[k] += v
    else:
        merged[k] = v


sorted_dict = {k: merged[k] for k in sorted(merged)}

print(sorted_dict)
```

```python
nums = list(map(int, input().split()))

k = int(input())

pairs = []

for i in range(len(nums)):
    for j in range(i+1, len(nums)):
        if nums[i] + nums[j] == k:
            pairs.append((nums[i], nums[j]))


pairs.sort()

print(pairs)
```

```python
def find_most_frequent(nums):
    if not nums:
        return None
    count_dict = {}
    for x in nums:
        count_dict[x] = count_dict.get(x, 0) + 1
    max_freq = max(count_dict.values())
    candidates = [k for k, v in count_dict.items() if v == max_freq]
    return min(candidates)
input_list = list(map(int, input().split()))
result = find_most_frequent(input_list)
print(f"{result}")
```

```python
def tim_vi_tri_k(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i + 1
nums = list(map(int, input().split()))
k = int(input())
print(tim_vi_tri_k(nums, k))
```
