```python
try:
    a = int(input())
    b = int(input())
    result = a / b
    print(f"{result:.2f}")
except ZeroDivisionError:
    print("Loi chia khong")
except ValueError:
    print("Loi nhap lieu")
```

```python
try:
    n = int(input())
    my_list = [int(input()) for _ in range(n)]
    print(my_list[n])
except IndexError:
    print("Truy cap chi so sai")
except ValueError:
    print("Loi nhap lieu")
```

```python
def is_happy(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(i)**2 for i in str(n))
    return n == 1

try:
    a = int(input())
    if a <= 0:
        print("NO")
    else:
        print("YES" if is_happy(a) else "NO")
except ValueError:
    print("Loi nhap lieu")
```

```python
data = input().split()
result = []
for item in data:
    try:
        result.append(int(item))
    except ValueError:
        print(f"Canh bao: '{item}' khong the chuyen ve int")
print(result)
```

```python
try:
    x = float(input())
    if x < 0:
        raise ValueError("So am")
    print(f"{x**0.5:.2f}")
except ValueError as e:
    if str(e) == "So am":
        print("So am")
    else:
        print("Loi nhap lieu")
```

```python
try:
    val = input()
    age = int(val)
    if age < 0:
        print("Tuoi khong hop le")
    else:
        print(2025 - age)
except ValueError:
    print("Tuoi khong hop le")
```

```python
try:
    line1 = input().split()
    if not line1: exit()
    n = int(line1[0])
    arr = list(map(int, input().split()))

    if len(arr) == n and len(set(arr)) == len(arr):
        print(sum(arr))
    else:
        print("Mang khong hop le")
except (ValueError, IndexError):
    print("Mang khong hop le")
```

```python
import re

try:
    t = int(input())
    for _ in range(t):
        s = input()
        try:
            re.compile(s)
            print("True")
        except re.error:
            print("False")
except ValueError:
    pass
```
